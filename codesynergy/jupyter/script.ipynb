```python
# #  create range of monthly dates
# download_dates = pd.date_range(start='2023-01-01', end='2023-01-01' , freq="1h")

#  URL from Chrome DevTools Console
base_url = ("https://data.gharchive.org/2023-01-01-2.json.gz") #  add format option to year and month

# #  create list of remote URL from base URL
# list_of_url = [base_url.format(date.date(),date.hour) for date in download_dates]

# dfs = []
# for url in list_of_url:
#     print("url",url)
response = requests.get(base_url)
content = response.content
df = pd.read_json(io.BytesIO(content), lines=True, compression='gzip')
#     dfs.append(df_aux)

# len(dfs)
```

```python
df.info(memory_usage='deep')
```

```python
df
```

```python
df.fillna('', inplace=True)
```

```python
pandas_to_frame = df.to_dict('records')
```

```python
pandas_to_frame[0]
```

```python
for index, value in enumerate(pandas_to_frame):
        pandas_to_frame[index]["actor_id"] =  pandas_to_frame[index]["actor"]['id']
        pandas_to_frame[index]["actor_login"] =  pandas_to_frame[index]["actor"]['login']
        pandas_to_frame[index]["actor_display_login"] =  pandas_to_frame[index]["actor"]['display_login']
        pandas_to_frame[index]["actor_gravatar_id"] =  pandas_to_frame[index]["actor"]['gravatar_id']
        pandas_to_frame[index]["actor_url"] =  pandas_to_frame[index]["actor"]['url']
        pandas_to_frame[index]["actor_avatar_url"] =  pandas_to_frame[index]["actor"]['avatar_url']
```

```python
for index, value in enumerate(pandas_to_frame):
    pandas_to_frame[index]["repo_id"] =   pandas_to_frame[index]["repo"]['id']
    pandas_to_frame[index]["repo_name"] =   pandas_to_frame[index]["repo"]['name']
    pandas_to_frame[index]["repo_url"] =   pandas_to_frame[index]["repo"]['url']
```

```python
for index, value in enumerate(pandas_to_frame):
    if "id" in pandas_to_frame[index]["org"]:
        pandas_to_frame[index]["org_id"] =  pandas_to_frame[index]["org"]['id']
        pandas_to_frame[index]["org_login"] =  pandas_to_frame[index]["org"]['login']
        pandas_to_frame[index]["org_gravatar_id"] =  pandas_to_frame[index]["org"]['gravatar_id']
        pandas_to_frame[index]["org_url"] =  pandas_to_frame[index]["org"]['url']
        pandas_to_frame[index]["org_avatar_url"] =  pandas_to_frame[index]["org"]['avatar_url']
```

```python
df = pd.DataFrame(pandas_to_frame)
```
